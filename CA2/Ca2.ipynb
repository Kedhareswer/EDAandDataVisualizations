```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("StudentsPerformance.csv")

# Plot the boxplots for each subject
sns.boxplot(data=df[["math score", "reading score", "writing score"]])
plt.title("Boxplots of Students' Scores in Three Subjects")
plt.show()
```

### The interpretation of the graph is as follows:

A boxplot shows the distribution of a numerical variable by displaying five summary statistics: the minimum, the first quartile (Q1), the median (Q2), the third quartile (Q3), and the maximum. The box represents the interquartile range (IQR), which is the difference between Q3 and Q1. 

The median is shown as a line inside the box. The whiskers extend from the box to the minimum and maximum values, or to 1.5 times the IQR, whichever is smaller. Any data points beyond the whiskers are considered outliers and are shown as dots.


From the graph, we can see that the median scores for math, reading, and writing are around 66, 70, and 69, respectively. The IQRs for math and reading are similar, around 20, while the IQR for writing is slightly smaller, around 18. The minimum and maximum scores for math are 0 and 100, respectively, while for reading and writing they are both 17 and 100, respectively.


We can also see that there are some outliers in each subject. For math, there are four outliers below the lower whisker, which means that these students scored very low compared to their peers. For reading and writing, there are no outliers below the lower whisker, but there are some outliers above the upper whisker, which means that these students scored very high compared to their peers. 

The number of outliers for reading is four, while for writing it is six.
The boxplots can help us identify and analyze the outliers in each subject. 

We can investigate further why some students scored exceptionally low or high in a certain subject. For example, we can look at their demographic information or test preparation status to see if there are any patterns or correlations with their scores. We can also decide whether to keep or remove these outliers from our dataset depending on our research question or analysis objective.

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
 

# Plot the bar graph for gender vs average score
df["average score"] = (df["math score"] + df["reading score"] + df["writing score"]) / 3
sns.barplot(x="gender", y="average score", data=df)
plt.title("Bar Graph")
plt.show()

# Plot the box plot for test preparation course vs writing score
sns.boxplot(x="test preparation course", y="writing score", data=df)
plt.title("Box Plot")
plt.show()
```

### The interpretation of each graph is as follows:

The line chart shows the relationship between math score and reading score for each student. We can see that there is a positive correlation between the two variables, which means that students who scored higher in math also tended to score higher in reading, and vice versa. The line chart also shows some outliers, such as students who scored very low or very high in one subject but not in the other.


The bar graph shows the comparison of average score between male and female students. We can see that female students had a higher average score than male students, which means that female students performed better overall in all three subjects. The bar graph also shows the error bars, which indicate the standard deviation of the average score for each gender group. We can see that the error bars overlap slightly, which means that there is some variation within each group and that the difference between the groups is not very large.


The box plot shows the distribution of writing score for each test preparation course category: none or completed. We can see that students who completed the test preparation course had a higher median writing score than students who did not complete it, which means that completing the test preparation course had a positive effect on writing performance. The box plot also shows the interquartile range (IQR), which is the difference between the third quartile (Q3) and the first quartile (Q1). The IQR represents the middle 50% of the data. We can see that the IQR for both categories is similar, which means that there is no significant difference in the variability of writing scores within each category. The box plot also shows some outliers, such as students who scored very low or very high in writing regardless of their test preparation course status.

# 6. Briefly discuss about Stacked Bar Graphs with an example

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create a dataframe with sample data
df = pd.DataFrame({
    "Quarter": ["Q1", "Q2", "Q3", "Q4"],
    "Apple": [20, 25, 30, 35],
    "Banana": [15, 20, 25, 30],
    "Orange": [10, 15, 20, 25],
    "Grape": [5, 10, 15, 20]
})

# Plot the stacked bar graph
df.plot(kind="bar", x="Quarter", stacked=True)
plt.title("Stacked Bar Graph of Fruit Sales by Quarter")
plt.xlabel("Quarter")
plt.ylabel("Sales (in thousands)")
plt.legend(title="Fruit")
plt.show()
```

### The interpretation of the graph is as follows:

The graph shows the total sales of fruits in each quarter as well as the sales of each fruit within each quarter. The total sales are represented by the height of each bar, and the sales of each fruit are represented by the color-coded segments within each bar.

From the graph, we can see that the total sales increased in each quarter, from 50 thousand in Q1 to 110 thousand in Q4. We can also see that apple was the most sold fruit in each quarter, followed by banana, orange, and grape. The sales of each fruit also increased in each quarter, except for grape in Q2, which remained the same as in Q1.

The graph can help us compare the sales performance of different fruits across different quarters. We can also see how each fruit contributed to the total sales in each quarter. For example, we can see that apple accounted for 40% of the total sales in Q1 and Q2, but increased to 45% in Q3 and Q4. We can also see that grape accounted for 10% of the total sales in Q1 and Q2, but decreased to 9% in Q3 and Q4.
