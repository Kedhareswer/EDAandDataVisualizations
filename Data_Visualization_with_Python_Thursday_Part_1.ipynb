```python
# Scatter Plot

# Importing libraries
import matplotlib.pyplot as plt

# Creating the data
x = [5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 100, 86, 
    103, 87, 94, 78, 77, 85, 86]

# Setting the figure size 
plt.figure(figsize = (7, 5))

plt.scatter(x, y, label = 'Points', c = 'red')

# Set x and y axes labels
plt.xlabel('X values')
plt.ylabel('Y values')

plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
# Box Plot

# Importing numpy
import numpy as np

# Creating the data
data = np.random.normal(100, 20, 200)

# Creating the plot
plt.boxplot(data)
plt.xlabel('Variable')
plt.ylabel('Values')
plt.title('Boxplot')
plt.show()
```

```python
# Histogram

data = [1.7,1.8,2.0,2.2,2.2,2.3,2.4,2.5,2.5,2.5,2.6,2.6,2.8,
       2.9,3.0,3.1,3.1,3.2,3.3,3.5,3.6,3.7,4.1,4.1,4.2,4.3]

# Plot
plt.hist(data, range = (1, 4), bins = 8)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
