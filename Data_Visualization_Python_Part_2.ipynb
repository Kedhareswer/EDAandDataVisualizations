```python
# Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Topics to be covered today -
# (1) Cat Plot
# (2) Dist Plot
# (3) Pair Plot
# (4) Joint Plot
# (5) Bar Plot
# (6) Density Plot
```

```python
# Catplot
exercise = sns.load_dataset('exercise')
exercise.head()
```

```python
# cat Plot
sns.catplot(x = 'time', y = 'pulse', hue = 'kind', data = exercise)
```

```python
# hue : encodes the points with different colors w.r.t. target variable
```

```python
# distplot
df = sns.load_dataset('tips')
df.head()
```

```python
# distplot
sns.distplot(df['total_bill'], kde = False, color = 'red', bins = 10)
```

```python
# distplot
sns.distplot(df['total_bill'], kde = True, color = 'green', bins = 10)
```

```python
# Pair Plot
sns.pairplot(df, hue = 'sex', palette = 'coolwarm')
```

```python
# Pair Plot
sns.pairplot(df, hue = 'smoker', palette = 'coolwarm')
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df)
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde')
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
```

```python
# jointplot
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
```

```python
# kind can have hex, reg, kde values
# Default value for kind is scatter.
```

```python
# barplot
df = sns.load_dataset('titanic')
df = pd.DataFrame(df)
df.head()
```

```python
# barplot
sns.barplot(x = 'who', y = 'fare', data = df)
```

```python
# barplot
sns.barplot(x = 'fare', y = 'who', data = df)
```

```python
# Clustered Bar Chart
sns.barplot(x = 'who', y = 'fare', hue = 'class', data = df)
```

```python
# Clustered Bar Chart
sns.barplot(x = 'who', y = 'fare', hue = 'alive', data = df)
```

```python
# Clustered Bar Chart
sns.barplot(x = 'who', y = 'fare', hue = 'class', data = df, palette = 'Blues')
```

```python
# distplot
np.random.seed(7)
num = np.random.randn(700)
num = pd.Series(num, name = ' Number')

# Histogram Plot
sns.distplot(num, kde = False)
```

```python
# Density Plot
sns.distplot(num, kde = True)
```
