```python
def per_check(data):
    n = df.shape[0]
    drop_col = []
    for i in data.columns:
        x = data[i].isnull().sum()
        print(i,(x/n) * 100)
        if((x/n) * 100) > 40:
            drop_col.append(i)
```

```python
per_check(df)
```

```python
# Lets rename the columns and encode Yes/No to True or False cause this is frustating
# I have changed a few column names through excel
```

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    df[col] = df[col].map(yes_no)

df.head()
```

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in df.columns[3:9]:
    df[col] = df[col].map(movie_mapping)
```

```python
star_wars = df.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

```python

star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
# Lets drop the columns with more than 40% nulls
dd = star_wars.drop('Do you consider yourself to be a fan of the Expanded Universe?æ',axis=1)
```

```python
# Lets use Mean or Median to fill the null with less than 40% null and Greater than 5% nulls
dd['ranking_1'] = dd['ranking_1'].fillna(dd.ranking_1.median())
dd['ranking_2'] = dd['ranking_2'].fillna(dd.ranking_2.median())
dd['ranking_3'] = dd['ranking_3'].fillna(dd.ranking_3.median())
dd['ranking_4'] = dd['ranking_4'].fillna(dd.ranking_4.median())
dd['ranking_5'] = dd['ranking_5'].fillna(dd.ranking_5.median())
dd['ranking_6'] = dd['ranking_6'].fillna(dd.ranking_6.median())
```

```python
per_check(dd)
```

```python
dd['Do you consider yourself to be a fan of the Star Wars film franchise?']= dd['Do you consider yourself to be a fan of the Star Wars film franchise?'].fillna(True)
```

```python
object_columns = dd.select_dtypes(include=['object']).columns

# Fill missing values with mode for object data type columns
for column in object_columns:
    dd[column] = dd[column].fillna(dd[column].mode()[0])
```
