```python
da = pd.read_csv('bank 1.csv')
```

```python
mode_value = da['age'].mode()[0]
da['age'].fillna(mode_value, inplace=True)
```

```python
mode_value = da['contact'].mode()[0]
da['contact'].fillna(mode_value, inplace=True)
```

```python
median_value = da['duration'].median()  
da['duration'].fillna(median_value, inplace=True)
```

```python
dd = da.isnull().sum()
dd
```

```python
plt.figure(figsize=(5,5))

da.marital.value_counts(normalize=True).plot.barh(color=['blue', 'green', 'orange'])

plt.xlabel('Percentage', fontsize=15)

plt.ylabel('Status', fontsize=15)

plt.title('Martial Status', fontsize = 24)

custom_labels = ['Married', 'Single', 'Divorced']
handles = [plt.Rectangle((0,0),1,1, color='blue'),
           plt.Rectangle((0,0),1,1, color='green'),
           plt.Rectangle((0,0),1,1, color='orange')]
plt.legend(handles, custom_labels)

plt.show()
```
