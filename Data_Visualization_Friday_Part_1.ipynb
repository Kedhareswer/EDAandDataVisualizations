```python
# Scatter Plot 

# Importing matplotlib
import matplotlib.pyplot as plt 

# Declaring x and y values
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,100,86,103,87,94,78,77,85,86]

# Setting the figure size (in inches)
plt.figure(figsize = (7, 5))

# Plot
plt.scatter(x, y, label = 'Points', c = 'red')

# Setting the labels
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
# Line Plot

# Creating the data
year = [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
unemployment_rate = [9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3,10]

# Plot
plt.plot(year, unemployment_rate, c = 'green')
plt.title('Unemployment Rate vs Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
# Box Plot

# Importing numpy 
import numpy as np

# Creating the dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)

# Creating the plot
plt.boxplot(data)
plt.title('Boxplot')
plt.ylabel('Values')
plt.show()
```

```python
# Histogram 

# Creating the data
data = [1.7,1.8,2.0,2.2,2.2,2.3,2.4,2.5,2.5,2.5,2.6,2.6,2.8,
       2.9,3.0,3.1,3.1,3.2,3.3,3.5,3.6,3.7,4.1,4.1,4.2,4.3]

# Plot
plt.hist(data, range = (1, 4), bins = 8)
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title("Histogram")
plt.show()
```

```python
# Pie Chart

# Creating the data
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]

# Plot
plt.pie(sizes, labels = labels,
       autopct = '%1.2f%%', shadow = True)
plt.title('Pie Chart for composition of learners')
plt.show()
```
